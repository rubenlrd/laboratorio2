**Paso 8: elimminar cualquier otra fila que aun contenga valores nulos**

```python
df = df.dropna()
df
```

```python
df.isnull().sum()
```

**Paso 9. Convertir a formato fecha (fecha de pedido y fecha de envio)**

```python
df['Fecha del pedido'] = pd.to_datetime(df['Fecha del pedido'], dayfirst=True)
df
```

```python
df['Fecha de envío'] = pd.to_datetime(df['Fecha de envío'], dayfirst=True)
df.info()
```

**Paso 10:Eliminar columnas imnecesarias**

```python
df.drop(columns=['Id. de la fila', 'Id. del pedido'], axis=1, inplace=True)
df.info()
```

```python
df.drop(columns=['Id. del cliente', 'Id. del producto'], axis=1, inplace=True)
df.info()
```

**Visualización**

Se utilizo LinePlot para visualizar tipos de clientes por ventas

```python
# Ordeno por fecha de pedido
df_ordenado = df.sort_values(by='Fecha del pedido', ascending=True)
sns.lineplot(data=df_ordenado, x='Fecha del pedido', y='Ventas')
plt.title('Ventas Por Año')
plt.xlabel('Fecha de Pedido')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

ver

```python
sns.barplot(data=df, x='País/Región', y='Ventas', estimator='mean', ci=None)
plt.xticks(rotation=45)
plt.title('Ventas promedio por Pais - Región')
plt.xlabel('Pais/Región')
plt.ylabel('Ventas')
```

```python
sns.histplot(data=df, x='Subcategoría', bins=20, kde=True)
plt.xticks(rotation=45)
plt.title('Distribución de Subcategorias ')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.violinplot(data=df, x='Segmento', y='Ventas')
plt.title('Distribución de Ventas por Segmento')
plt.xlabel('Segmento')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
```

```python
sns.jointplot(data=df, x='Cantidad', y='Ventas', kind='scatter')
plt.title('Relación entre Ventas y Cantidad de productos por pedido')
plt.xlabel('Ventas')
plt.ylabel('Cantidad')
plt.show()
```

```python
boxplot = sns.boxplot(data=df, x='Región', y='Ventas')
plt.xticks(rotation=45)
plt.title('Distribución de Ventas por Región')
plt.xlabel('Región')
plt.ylabel('Ventas')
```

```python
boxplot = sns.boxplot(data=df, x='Categoría', y='Ventas')
plt.xticks(rotation=45)
plt.title('Distribución de Ventas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas')
```

```python
sns.lmplot(data=df, x='Ventas', y='Ganancia', hue='País/Región')
plt.title('Relación entre Ventas y Cantidad por Pais Región')
plt.xlabel('Ventas')
plt.ylabel('Cantidad')
```

```python
sns.pairplot(df)
plt.show()
```

```python
pivot = df.pivot_table(index='País/Región', columns='Categoría', values='Ventas', aggfunc='sum')
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title('Ventas por País/Región y Categoría')
plt.show()
```

**Se calcula la correlación**

```python
correlacion = df['Ventas'].corr(df['Ganancia'])
print ("Correlacion entre ventas y ganancias: ", correlacion)
```
